```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories=["A","B","C","D"]
val1=[4,5,3,6]
val2=[2,5,4,3]

x=np.arange(len(categories))



plt.figure(figsize=(10,6))
plt.bar(x,val1,color="skyblue",width=0.2)
plt.xticks(x,categories)
plt.xlabel("Categories")
plt.title("column chart")
plt.ylabel("Values")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.barh(categories,val1,color="orange")
plt.xlabel("Categories")
plt.title("column chart")
plt.ylabel("Values")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(x,val1,color="skyblue", label="values1")

plt.bar(x,val2,bottom=val1,color="lightgreen", label="values2")
plt.xticks(x,categories)
plt.xlabel("Categories")
plt.title("Stacked BAr Graph")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.barh(x,val1,color="skyblue", label="values1",height=0.6)
plt.barh(x,val2,height=0.6,color="lightgreen", label="values2")
plt.xticks(x,categories)
plt.xlabel("Categories")
plt.title("Stacked BAr Graph")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
x=np.arange(1,6)
y1=np.array([2,4,6,5,7])
y2=np.array([1,3,4,6,8])

plt.figure(figsize=(10,6))
plt.stackplot(x,y1,y2,labels=["Series1","Series2"],alpha=0.6)
plt.xlabel("x-axis")
plt.title("Area Chart")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
fig,ax1=plt.subplots(figsize=(8,5))
ax2=ax1.twinx()
ax1.bar(x,y1,color="skyblue",label="bar chart")
ax2.plot(x,y2,color="orange",marker="o",label="line data")

ax1.set_xlabel("X-axis")
ax1.set_ylabel("bar chart")
ax2.set_ylabel("line chart")
plt.title("line values")
plt.show()
```

```python
widths = [0.5, 1.0, 1.5, 1.0, 2.0]
val = [3, 4, 2, 6, 4]

lefts = np.cumsum([0] + widths[:-1])
plt.figure(figsize=(8, 5))
plt.bar(lefts, val, width=widths, color="lightgreen", edgecolor="black")
plt.title("Meeko Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
