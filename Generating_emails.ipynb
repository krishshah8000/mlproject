```python
import csv
import random

def generate_realistic_emails(filename="realistic_emails_2000.csv", count=2000):
    # Contextual Scenarios
    scenarios = {
        "High": [
            {"topic": "Data Loss", "desc": "Files are missing, servers are wiped, or backups are failing."},
            {"topic": "Legal/Compliance", "desc": "Audit deadlines, contract breaches, or regulatory fines."},
            {"topic": "Customer Crisis", "desc": "Big clients threatening to leave or major payment failures."},
            {"topic": "Security", "desc": "Breaches, suspicious logins, or exposed API keys."},
            {"topic": "Production Outage", "desc": "Site down, checkout broken, or major API latency."}
        ],
        "Medium": [
            {"topic": "Project Progress", "desc": "Requesting updates, sending drafts, or asking for feedback."},
            {"topic": "Meetings", "desc": "Scheduling syncs, moving times, or sending agendas."},
            {"topic": "Internal Questions", "desc": "Clarifying a task, asking about a policy, or tech questions."},
            {"topic": "Resource Needs", "desc": "Ordering equipment, hiring updates, or software access."}
        ],
        "Low": [
            {"topic": "Social/Culture", "desc": "Office lunch, happy hours, or team building."},
            {"topic": "General Info", "desc": "Newsletters, new office decor, or parking updates."},
            {"topic": "Learning/Leisure", "desc": "Shared articles, optional training, or industry news."},
            {"topic": "Admin/FYI", "desc": "Holiday reminders, generic internal announcements."}
        ]
    }

    # Sentence fragments to build more human-like bodies
    greetings = ["Hi team,", "Hey everyone,", "Hello,", "Hi,", "Morning,"]
    closings = ["Regards,", "Thanks,", "Best,", "Let me know,", "Cheers,"]

    with open(filename, mode='w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['subject', 'body', 'priority'])

        generated_pairs = set()

        while len(generated_pairs) < count:
            priority = random.choice(["High", "Medium", "Low"])
            scenario = random.choice(scenarios[priority])

            # Create varied content based on scenario
            if priority == "High":
                subjs = [
                    f"Issue with {scenario['topic']}",
                    f"Immediate action needed for {scenario['topic']}",
                    f"Major problem: {scenario['topic']}",
                    f"{scenario['topic']} Alert"
                ]
                bodies = [
                    f"I've just noticed that {scenario['desc']} This is causing a complete block for the team. We need this fixed before the morning meeting.",
                    f"The client is furious because {scenario['desc']} If we don't resolve this in the next few hours, we risk losing the account.",
                    f"Our monitoring system triggered an alert: {scenario['desc']} I've checked the logs and it looks critical."
                ]

            elif priority == "Medium":
                subjs = [
                    f"Quick question about {scenario['topic']}",
                    f"Update regarding {scenario['topic']}",
                    f"Following up on {scenario['topic']}",
                    f"{scenario['topic']} sync"
                ]
                bodies = [
                    f"I'm currently working on the {scenario['topic']} and wanted to check {scenario['desc']} Do you have a moment to review this week?",
                    f"Can you provide an update on the {scenario['topic']}? We're trying to finalize the plan for next month.",
                    f"Regarding the {scenario['topic']}, {scenario['desc']} I've attached the latest version for your comments."
                ]

            else: # Low
                subjs = [
                    f"FYI: {scenario['topic']}",
                    f"Thinking about {scenario['topic']}",
                    f"Random share: {scenario['topic']}",
                    f"Note on {scenario['topic']}"
                ]
                bodies = [
                    f"I found some info about {scenario['topic']} and thought you might find it interesting. {scenario['desc']}",
                    f"Just wanted to let everyone know that {scenario['desc']} No need to reply, just for your information.",
                    f"Hope you're all having a good week. {scenario['desc']} Check the link when you have some free time."
                ]

            subject = random.choice(subjs)
            # Add human "noise" to the body
            body = f"{random.choice(greetings)}\n\n{random.choice(bodies)}\n\n{random.choice(closings)}"

            if (subject, body) not in generated_pairs:
                writer.writerow([subject, body, priority])
                generated_pairs.add((subject, body))

    print(f"Generated {count} highly realistic records in {filename}")

generate_realistic_emails()
```
