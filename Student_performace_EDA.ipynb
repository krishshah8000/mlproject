```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Show Data

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head(10)
```

Cleaning Process

```python
df.info()
```

Statistics of data

```python
df.describe()
```

Distribution of Male and Female Students

```python
sns.countplot(x='gender', data=df)
plt.title('Distribution of Male and Female Students')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Distribution of Students by Race/Ethnicity

```python
race_counts = df['race/ethnicity'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(race_counts, labels=race_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Students by Race/Ethnicity')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Distribution of Parental Level of Education

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='parental level of education', data=df, order=df['parental level of education'].value_counts().index)
plt.title('Distribution of Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Distribution of Test Preparation Course Completion

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='test preparation course', data=df)
plt.title('Distribution of Test Preparation Course Completion')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.show()
```

Distribution of Lunch Types

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='lunch', data=df)
plt.title('Distribution of Lunch Types')
plt.xlabel('Lunch Type')
plt.ylabel('Count')
plt.show()
```

Reading Scores by Parental Level of Education

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='parental level of education', y='reading score', data=df, order=df['parental level of education'].value_counts().index)
plt.title('Reading Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Reading Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Writing Score Distribution by Parental Level of Education

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='parental level of education', y='writing score', data=df, order=df['parental level of education'].value_counts().index)
plt.title('Writing Score Distribution by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Writing Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Math Score Distribution by Parental Level of Education

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='parental level of education', y='math score', data=df, order=df['parental level of education'].value_counts().index)
plt.title('Math Score Distribution by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Math Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Average Scores by Parental Level of Education

```python
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

plt.figure(figsize=(12, 6))
sns.boxplot(x='parental level of education', y='average score', data=df, order=df['parental level of education'].value_counts().index)
plt.title('Average Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Average Scores by Gender

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='average score', data=df)
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.show()
```

Correlation Matrix: Math, Reading, and Writing Scores

```python
correlation_matrix_scores = df[['math score', 'reading score', 'writing score']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_scores, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix: Math, Reading, and Writing Scores')
plt.show()
```

Average Scores by Race/Ethnicity

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='race/ethnicity', y='average score', data=df)
plt.title('Average Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Score')
plt.show()
```

Gender wise statistics data of each score

```python
display(df.groupby('gender')[['math score', 'reading score', 'writing score']].describe())
```

Scatter Plot of Math Score vs. Reading Score

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='math score', y='reading score', data=df)
plt.title('Scatter Plot of Math Score vs. Reading Score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

Scatter Plot of Math Score vs. Writing Score

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='math score', y='writing score', data=df)
plt.title('Scatter Plot of Math Score vs. Writing Score')
plt.xlabel('Math Score')
plt.ylabel('Writing Score')
plt.show()
```

Average score who have complete the test preparation course or not

```python
display(df.groupby('test preparation course')['average score'].mean())
```

Average Math, Reading, and Writing Scores by Race/Ethnicity

```python
race_scores = df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean().reset_index()

race_scores_melted = race_scores.melt(id_vars='race/ethnicity', var_name='score_type', value_name='average_score')

plt.figure(figsize=(12, 6))
sns.barplot(x='race/ethnicity', y='average_score', hue='score_type', data=race_scores_melted)
plt.title('Average Math, Reading, and Writing Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
