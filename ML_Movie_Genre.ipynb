```python
import json

def extract_cast_names(cast_str):
    try:
        # Ensure the input is a string before attempting to load
        if not isinstance(cast_str, str):
            return ''

        cast_list = json.loads(cast_str)
        names = [member['name'] for member in cast_list if 'name' in member]
        return ', '.join(names)
    except json.JSONDecodeError:
        return '' # Return an empty string if JSON parsing fails
    except Exception as e:
        # Handle other potential errors during processing
        print(f"Error processing cast string: {cast_str} - {e}")
        return ''

movieCredit['cast'] = movieCredit['cast'].apply(extract_cast_names)
print("Extracted cast names and updated 'cast' column in movieCredit DataFrame.")
print(movieCredit.head())
```

```python
import json

def extract_department_names(crew_str):
    try:
        # Ensure the input is a string before attempting to load
        if not isinstance(crew_str, str):
            return []

        crew_list = json.loads(crew_str)
        departments = [member['department'] for member in crew_list if 'department' in member]
        return departments
    except json.JSONDecodeError:
        return [] # Return an empty list if JSON parsing fails
    except Exception as e:
        # Handle other potential errors during processing
        print(f"Error processing crew string: {crew_str} - {e}")
        return []

movieCredit['crew_departments'] = movieCredit['crew'].apply(extract_department_names)
print("Extracted crew department names and added 'crew_departments' column in movieCredit DataFrame.")
print(movieCredit.head())
```

```python
all_departments = movieCredit['crew_departments'].explode().dropna().unique()
print("Unique Department Names:")
print(sorted(all_departments))
```

```python
import json

def extract_directors(crew_str):
    try:
        if not isinstance(crew_str, str):
            return ''
        crew_list = json.loads(crew_str)
        directors = [member['name'] for member in crew_list if member.get('department') == 'Directing' and member.get('job') == 'Director']
        return ', '.join(directors)
    except json.JSONDecodeError:
        return ''
    except Exception as e:
        print(f"Error processing crew string: {crew_str} - {e}")
        return ''

movieCredit['directing_crew'] = movieCredit['crew'].apply(extract_directors)
print("Extracted directing crew names and added 'directing_crew' column in movieCredit DataFrame.")
print(movieCredit.head())
```

```python
import json

def extract_writers(crew_str):
    try:
        if not isinstance(crew_str, str):
            return ''
        crew_list = json.loads(crew_str)
        writers = [member['name'] for member in crew_list if member.get('department') == 'Writing']
        return ', '.join(writers)
    except json.JSONDecodeError:
        return ''
    except Exception as e:
        print(f"Error processing crew string: {crew_str} - {e}")
        return ''

movieCredit['writing_crew'] = movieCredit['crew'].apply(extract_writers)
print("Extracted writing crew names and added 'writing_crew' column in movieCredit DataFrame.")
print(movieCredit.head())
```

```python
def get_unique_names(names_str):
    if not isinstance(names_str, str) or not names_str.strip():
        return ''
    names = [name.strip() for name in names_str.split(',')]
    unique_names = sorted(list(set(names))) # Sort for consistent output
    return ', '.join(unique_names)

movieCredit['cast'] = movieCredit['cast'].apply(get_unique_names)
print("Applied get_unique_names function to 'cast' column.")
print(movieCredit.head())
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv', engine='python', skipinitialspace=True)
print("Loaded 'tmdb_5000_movies.csv' into 'movies' DataFrame.")
print(movies.head())
```

```python
movies = movies[['genres', 'id', 'keywords', 'original_title', 'overview', 'production_companies']]
print("Selected specified columns from the 'movies' DataFrame.")
print(movies.head())
```

```python
import json

def extract_names_from_json(json_str):
    try:
        if not isinstance(json_str, str):
            return ''
        list_of_dicts = json.loads(json_str)
        names = [d['name'] for d in list_of_dicts if 'name' in d]
        return ', '.join(names)
    except (json.JSONDecodeError, TypeError):
        return ''
    except Exception as e:
        print(f"Error processing JSON string: {json_str} - {e}")
        return ''

movies['genres'] = movies['genres'].apply(extract_names_from_json)
print("Extracted names from 'genres' column.")
movies['keywords'] = movies['keywords'].apply(extract_names_from_json)
print("Extracted names from 'keywords' column.")
movies['production_companies'] = movies['production_companies'].apply(extract_names_from_json)
print("Extracted names from 'production_companies' column.")

print("\nUpdated 'movies' DataFrame head:")
print(movies.head())
```

```python
movies.drop('production_companies', axis=1, inplace=True)
print("Dropped 'production_companies' column from the movies DataFrame.")
print(movies.head())
```

```python
movieCredit['movie_id'] = pd.to_numeric(movieCredit['movie_id'], errors='coerce')
movieCredit.dropna(subset=['movie_id'], inplace=True)
movieCredit['movie_id'] = movieCredit['movie_id'].astype(int)
merged_movies = pd.merge(movieCredit, movies, left_on='movie_id', right_on='id', how='inner')
print("Merged 'movieCredit' and 'movies' DataFrames into 'merged_movies'.")
print(merged_movies.head())
```

```python
merged_movies.drop(columns=['id', 'original_title'], inplace=True)
merged_movies.rename(columns={'title': 'movie_name'}, inplace=True)
print("Removed 'id' and 'original_title' columns and renamed 'title' to 'movie_name'.")
print(merged_movies.head())
```

```python
duplicate_rows = merged_movies[merged_movies.duplicated()]

if not duplicate_rows.empty:
    print(f"Found {len(duplicate_rows)} duplicate rows in `merged_movies`:")
    display(duplicate_rows)
else:
    print("No duplicate rows found in `merged_movies`.")
```

```python
merged_movies['overview'] = merged_movies['overview'].fillna('')
print("Filled missing values in 'overview' column with an empty string.")
print(merged_movies.isnull().sum())
```

```python
merged_movies['overview'].fillna('', inplace=True)
print("Filled missing values in 'overview' column with an empty string.")
print(merged_movies.isnull().sum())
```

```python
def remove_spaces_from_names(names_str):
    if not isinstance(names_str, str) or not names_str.strip():
        return ''
    names = [name.replace(' ', '') for name in names_str.split(',')]
    return ','.join(names)

columns_to_process = ['cast', 'directing_crew', 'writing_crew', 'genres', 'keywords']

for col in columns_to_process:
    merged_movies[col] = merged_movies[col].apply(remove_spaces_from_names)
    print(f"Removed spaces from names in '{col}' column.")

print("\nUpdated 'merged_movies' DataFrame head:")
print(merged_movies.head())
```

```python
merged_movies['tags'] = merged_movies.apply(lambda row:
    ' '.join(
        (row['genres'].split(',') * 3 if row['genres'] else []) +
        (row['keywords'].split(',') * 3 if row['keywords'] else []) +
        (row['cast'].split(',') * 2 if row['cast'] else []) +
        (row['directing_crew'].split(',') * 2 if row['directing_crew'] else []) +
        (row['writing_crew'].split(',') * 1 if row['writing_crew'] else []) +
        (row['overview'].split() if row['overview'] else [])
    ),
    axis=1
)

print("Updated 'tags' column with proper spacing between elements.")
print(merged_movies[['movie_name', 'tags']].head())
```

```python
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Download NLTK stopwords if not already downloaded
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')
print("NLTK stopwords are ready.")
```

```python
merged_movies['tags'] = merged_movies['tags'].str.lower()
print("Converted 'tags' column to lowercase.")
print(merged_movies.head())
```

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return ' '.join([word for word in str(text).split() if word not in stop_words])

merged_movies['tags'] = merged_movies['tags'].apply(remove_stopwords)
print("Removed stopwords from 'tags' column.")
print(merged_movies.head())
```

```python
tfidf = TfidfVectorizer(max_features=5000)
tfidf_matrix = tfidf.fit_transform(merged_movies['tags'])

print("TF-IDF vectorization applied to 'tags' column.")
print(f"Shape of TF-IDF matrix: {tfidf_matrix.shape}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix)
print("Cosine similarity matrix calculated.")
print(f"Shape of cosine similarity matrix: {cosine_sim.shape}")
```

```python
def recommend_movies(movie_title, merged_movies_df, cosine_sim_matrix, num_recommendations=5):
    # Get the index of the movie that matches the title, allowing for partial, case-insensitive matching
    try:
        # Find all movies that partially match the title (case-insensitive)
        matching_movies = merged_movies_df[merged_movies_df['movie_name'].str.contains(movie_title.strip(), case=False, na=False)]
        if matching_movies.empty:
            return f"Movie '{movie_title}' not found in the dataset."
        # Get the index of the first matching movie
        idx = matching_movies.index[0]
    except IndexError:
        return f"Movie '{movie_title}' not found in the dataset."

    # Get the cosine similarity scores for that movie with all other movies
    sim_scores = list(enumerate(cosine_sim_matrix[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the `num_recommendations` most similar movies (excluding itself)
    sim_scores = sim_scores[1:num_recommendations+1]

    # Get the movie indices and their similarity scores
    movie_indices = [i[0] for i in sim_scores]
    movie_similarities = [i[1] for i in sim_scores]

    # Return the top `num_recommendations` similar movies with their scores
    recommendations = pd.DataFrame({
        'Recommended Movie': merged_movies_df['movie_name'].iloc[movie_indices].values,
        'Similarity Score': movie_similarities
    })
    return recommendations
```
