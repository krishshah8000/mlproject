```python
import nltk
from nltk.stem import WordNetLemmatizer, PorterStemmer
import pandas as pd

# Download necessary NLTK data
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# Initialize lemmatizer and stemmer
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

# List of 15 words to process
words = [
    "running", "flies", "ponies", "feet", "cats",
    "better", "rocks", "corpora", "studies", "studying",
    "cried", "crying", "leaves", "generous", "computation"
]

# Apply lemmatization and stemming
results = []
for word in words:
    results.append({
        "Original": word,
        "Lemmatized (WordNet)": lemmatizer.lemmatize(word, pos='v'),  # Using 'v' for verb context
        "Stemmed (Porter)": stemmer.stem(word)
    })

# Create a DataFrame for comparison
df_comparison = pd.DataFrame(results)
display(df_comparison)
```

### Comparison Insights

*   **Lemmatization (WordNet)**: Aims to find the dictionary base form (lemma) of a word. It uses a vocabulary and morphological analysis. For example, 'flies' becomes 'fly' and 'feet' becomes 'foot' (if pos is not specified, it defaults to noun; here I used verb context for demonstration).
*   **Stemming (Porter)**: Uses heuristic rules to chop off the ends of words. It often results in non-dictionary words. For example, 'studies' becomes 'studi' and 'computation' becomes 'comput'.
