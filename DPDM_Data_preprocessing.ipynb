```python
import pandas as pd
```

```python
df = pd.read_csv("/content/customer.csv")
df
```

```python
missing_values = df.isnull().sum()
columns_with_missing_values = missing_values[missing_values > 0]

if not columns_with_missing_values.empty:
    print("Columns with missing values:")
    display(columns_with_missing_values)
else:
    print("No columns with missing values found.")
```

```python
df.info()
```

```python
duplicate_rows = df[df.duplicated()]

if not duplicate_rows.empty:
    print(f"Number of duplicate rows: {len(duplicate_rows)}")
    print("First 5 duplicate rows:")
    display(duplicate_rows.head())
else:
    print("No duplicate rows found.")
```

```python
# Identify categorical columns (object dtype)
categorical_cols = df.select_dtypes(include='object').columns

print("Unique values for each categorical field:")
for col in categorical_cols:
    print(f"\n--- Column: {col} ---")
    display(df[col].unique())
```

```python
df['EffectiveToDate'] = pd.to_datetime(df['EffectiveToDate'], format='mixed', errors='coerce').dt.strftime('%d/%m/%Y')
print("Converted 'EffectiveToDate' column to DD/MM/YYYY format:")
display(df[['Customer', 'EffectiveToDate']].head())
```

```python
if 'Unnamed: 0' in df.columns:
    df.drop('Unnamed: 0', axis=1, inplace=True)
    print("'Unnamed: 0' column removed successfully.")
    display(df.head())
else:
    print("'Unnamed: 0' column not found in the DataFrame.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['CustomerLifetimeValue'], bins=50, kde=True)
plt.title('Distribution of Customer Lifetime Value')
plt.xlabel('Customer Lifetime Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

categorical_cols = df.select_dtypes(include='object').columns.tolist()

# Exclude 'Customer' (high cardinality) and 'EffectiveToDate' (already handled as date)
columns_to_plot = [col for col in categorical_cols if col not in ['Customer', 'EffectiveToDate']]

print("Generating bar charts for categorical columns in a 2-column layout:")

# Calculate the number of rows needed for a 2-column layout
num_columns = 2
num_rows = math.ceil(len(columns_to_plot) / num_columns)

fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, 6 * num_rows))
axes = axes.flatten() # Flatten the 2D array of axes for easy iteration

for i, col in enumerate(columns_to_plot):
    ax = axes[i] # Get the current subplot axis
    value_counts = df[col].value_counts()
    sns.barplot(x=value_counts.index, y=value_counts.values, hue=value_counts.index, palette='viridis', legend=False, ax=ax)

    # Add labels to each bar
    for index, value in enumerate(value_counts.values):
        ax.text(index, value, str(value), ha='center', va='bottom', fontsize=9)

    ax.set_title(f'Distribution of {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', rotation=45) # Removed ha='right'

# Hide any unused subplots
for j in range(len(columns_to_plot), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
