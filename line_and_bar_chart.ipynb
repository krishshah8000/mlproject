```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
y2 = [1, 3, 5, 7]

plt.plot(x, y1)
plt.plot(x, y2)
plt.title("Multiple Line Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
val1 = [10, 20, 30]
val2 = [15, 18, 25]

plt.bar(x - 0.2, val1, width=0.4)
plt.bar(x + 0.2, val2, width=0.4)

plt.title("Grouped Bar Chart")
plt.show()
```
