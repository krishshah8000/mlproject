## Convert Duration to Total Minutes

### Subtask:
Parse the 'Duration' column to extract hours and minutes, then calculate the total duration in minutes and store it in a new column named 'duration_total_minutes'. This will handle cases with only hours, only minutes, or both.


**Reasoning**:
To convert the 'Duration' column into total minutes, I will first define a function to parse the duration strings and extract hours and minutes, then apply this function to create the 'duration_total_minutes' column as instructed.



```python
def convert_duration_to_minutes(duration):
    total_minutes = 0
    if 'h' in duration:
        parts = duration.split('h')
        total_minutes += int(parts[0].strip()) * 60
        if 'm' in parts[1]:
            total_minutes += int(parts[1].replace('m', '').strip())
    elif 'm' in duration:
        total_minutes += int(duration.replace('m', '').strip())
    return total_minutes

df['duration_total_minutes'] = df['Duration'].apply(convert_duration_to_minutes)
print("Created 'duration_total_minutes' column:")
print(df[['Duration', 'duration_total_minutes']].head())
```

**Reasoning**:
Now that the 'duration_total_minutes' column has been created, I will remove the original 'Duration' column as instructed. Then, I will display the DataFrame's information and the first few rows to verify the changes.



```python
df.drop('Duration', axis=1, inplace=True)
print("Original 'Duration' column dropped.")
df.info()
df.head()
```

## Convert 'Total_Stops' to numerical

### Subtask:
Convert the 'Total_Stops' column from string format to a numerical integer column. 'non-stop' should be converted to `0`, '1 stop' to `1`, and so on. Also, handle the single null value by filling it with `0` (assuming 'non-stop').

```python
def convert_total_stops(stops):
    if pd.isna(stops):
        return 0  # Assuming a NaN value means no stops
    if stops == 'non-stop':
        return 0
    return int(stops.split(' ')[0])

df['Total_Stops'] = df['Total_Stops'].apply(convert_total_stops)
print("Converted 'Total_Stops' column:")
print(df[['Total_Stops']].head())
df.info()
```

## Applying Label Encoding for Demonstration

**Note**: The columns 'Airline', 'Source', 'Destination', and 'Additional_Info' in your main `df` have already been one-hot encoded. To demonstrate Label Encoding on these original columns, I will create a new temporary DataFrame (`df_le`) by reloading the raw data and applying prior preprocessing steps (like converting 'Total_Stops' and extracting date components) before applying Label Encoding. This ensures your current `df` remains unchanged.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Create a fresh DataFrame copy to demonstrate Label Encoding
df_le = pd.read_excel("/content/Data_Train.xlsx")

# --- Reapply previous preprocessing steps to df_le ---

# Handle Dep_Time
df_le['Dep_Time'] = pd.to_datetime(df_le['Dep_Time'], format='%H:%M')
df_le['dep_hour'] = df_le['Dep_Time'].dt.hour
df_le['dep_min'] = df_le['Dep_Time'].dt.minute
df_le.drop('Dep_Time', axis=1, inplace=True)

# Handle Arrival_Time
df_le['Arrival_Time'] = pd.to_datetime(df_le['Arrival_Time'], errors='coerce')
df_le['arr_hour'] = df_le['Arrival_Time'].dt.hour
df_le['arr_min'] = df_le['Arrival_Time'].dt.minute
df_le.drop('Arrival_Time', axis=1, inplace=True)

# Handle Duration
def convert_duration_to_minutes(duration):
    total_minutes = 0
    if 'h' in duration:
        parts = duration.split('h')
        total_minutes += int(parts[0].strip()) * 60
        if 'm' in parts[1]:
            total_minutes += int(parts[1].replace('m', '').strip())
    elif 'm' in duration:
        total_minutes += int(duration.replace('m', '').strip())
    return total_minutes
df_le['duration_total_minutes'] = df_le['Duration'].apply(convert_duration_to_minutes)
df_le.drop('Duration', axis=1, inplace=True)

# Handle Total_Stops
def convert_total_stops(stops):
    if pd.isna(stops):
        return 0  # Assuming a NaN value means no stops
    if stops == 'non-stop':
        return 0
    return int(stops.split(' ')[0])
df_le['Total_Stops'] = df_le['Total_Stops'].apply(convert_total_stops)

# Handle Date_of_Journey
df_le['Date_of_Journey'] = pd.to_datetime(df_le['Date_of_Journey'], format='%d/%m/%Y')
df_le['journey_day'] = df_le['Date_of_Journey'].dt.day
df_le['journey_month'] = df_le['Date_of_Journey'].dt.month
df_le['journey_year'] = df_le['Date_of_Journey'].dt.year
df_le.drop('Date_of_Journey', axis=1, inplace=True)

# Drop 'Route' column as it was dropped before one-hot encoding
df_le.drop('Route', axis=1, inplace=True)

# --- Apply LabelEncoder to the specified columns ---
label_encoder = LabelEncoder()

for col in ['Airline', 'Source', 'Destination', 'Additional_Info']:
    df_le[col] = label_encoder.fit_transform(df_le[col])

print("DataFrame after Label Encoding:")
df_le.info()
df_le.head()
```
